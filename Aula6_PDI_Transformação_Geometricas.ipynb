```python
(Altura, Largura) = geometricaModificada2.shape[:2] 
centro = (Largura // 2, Altura // 2)
M = cv2.getRotationMatrix2D(centro, -90, 1.0)
img_rotacionada = cv2.warpAffine(geometricaModificada2, M, (Largura, Altura))
plt.imshow(img_rotacionada)
plt.title("Imagem Rotacionada")
plt.show()
```
