# Importação e Segmentando as imagens - Vanessa

##Imagem 1

```python
from google.colab import files
uploaded = files.upload()
```

```python
img2 = cv2.imread('TCGA_CS_4941_19960909_16.tif', cv2.IMREAD_UNCHANGED)
```

```python
plt.imshow(img2)
plt.title('Imagem BGR')
```

```python
grayTwo = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grayTwo, cmap='gray')
plt.title('Imagem BGR gray')
```

```python
grayTwo2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(grayTwo2)
plt.title('Imagem RGB')
```

```python
grayTwo3 = cv2.cvtColor(grayTwo2,cv2.COLOR_RGB2GRAY)
plt.imshow(grayTwo3, cmap='gray')
plt.title('Imagem BGR gray')
```

```python
METODO = cv2.THRESH_BINARY_INV
ret, thresh = cv2.threshold(grayTwo3,80,255,METODO)
fig = plt.figure(figsize=(15,15))

ax1 = fig.add_subplot(221)
plt.imshow(img2, cmap='gray')
plt.title('Imagem Original')

ax1 = fig.add_subplot(222)
plt.imshow(thresh, cmap='gray')
plt.title('Imagem Segmentada')

plt.show
```

```python
ret, thresh = cv2.threshold(grayTwo3,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
plt.imshow(thresh, cmap='gray')
plt.title('Imagem Binária com OTSU')
```

```python
img3 = cv2.imread('TCGA_CS_4941_19960909_14.tif', cv2.IMREAD_UNCHANGED)
```

```python
plt.imshow(img3)
plt.title('Imagem BGR')
```

```python
grayThree = cv2.cvtColor(img3,cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(grayThree,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
```

```python
grayThree1 = cv2.cvtColor(img3,cv2.COLOR_RGB2GRAY)
plt.imshow(img3, cmap='gray')
plt.title('Imagem RGB')
```

```python
imgEqualizada = cv2.equalizeHist(grayThree)
```

```python
fig5 = plt.figure(figsize=(10,10))
plt.subplots_adjust(left=0.125, bottom=0.1, right=1.5, top=1,wspace=0.33, hspace=0.2)
x1_plt = fig5.add_subplot(1, 2, 1)
x1_plt.hist(img3.ravel(),256,[0,256])
x2_plt = fig5.add_subplot(1, 2, 2)
x2_plt.hist(imgEqualizada.ravel(),256,[0,256])
plt.title('Histograma da equalização')
plt.show()
```

```python
ret, thresh = cv2.threshold(imgEqualizada,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
plt.imshow(thresh, cmap='gray')
plt.title('Imagem Binária com OTSU e qualizada')
```

```python
METODO = cv2.THRESH_BINARY_INV
ret, thresh3 = cv2.threshold(imgEqualizada,80,255,METODO)
fig = plt.figure(figsize=(15,15))

ax1 = fig.add_subplot(221)
plt.imshow(img3, cmap='gray')
plt.title('Imagem Original')

ax1 = fig.add_subplot(222)
plt.imshow(thresh3, cmap='gray')
plt.title('Imagem Segmentada Equalizada')

plt.show
```

## THRESH - Testes com outros métodos de segmentação de imagem

MASK

```python
graySix = cv2.cvtColor(img6,cv2.COLOR_BGR2GRAY)
ret, thresh6 = cv2.threshold(graySix,100,255,cv2.THRESH_BINARY_INV+cv2.THRESH_MASK)
plt.imshow(thresh6,cmap='gray')
```

TOZERO

```python
graySix = cv2.cvtColor(img6,cv2.COLOR_BGR2GRAY)
ret, thresh6 = cv2.threshold(graySix,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_TOZERO)
plt.imshow(thresh6,cmap='gray')
```

TOZERO INV

```python
graySix = cv2.cvtColor(img6,cv2.COLOR_BGR2GRAY)
ret, thresh6 = cv2.threshold(graySix,80,255,cv2.THRESH_BINARY_INV+cv2.THRESH_TOZERO_INV)
plt.imshow(thresh6,cmap='gray')
```

TRIANGLE

```python
graySix = cv2.cvtColor(img6,cv2.COLOR_BGR2GRAY)
ret, thresh6 = cv2.threshold(graySix,100,255,cv2.THRESH_BINARY_INV+cv2.THRESH_TRIANGLE)
plt.imshow(thresh6,cmap='gray')
```

TRUNC

```python
graySix = cv2.cvtColor(img6,cv2.COLOR_BGR2GRAY)
ret, thresh6 = cv2.threshold(graySix,20,255,cv2.THRESH_BINARY_INV+cv2.THRESH_TRUNC)
plt.imshow(thresh6,cmap='gray')
```
